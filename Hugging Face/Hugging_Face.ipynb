```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis",model= "tabularisai/multilingual-sentiment-analysis")
result = classifier("I am extremely disappointed with the service I received. The staff was rude, and the experience was terrible.")
```

```python
print(result)
```

```python
classifier = pipeline("sentiment-analysis")
task = ["I have a meeting at 3 PM tomorrow.",
        "The weather is cloudy today.",
        "She is wearing a blue dress."]
result1 = classifier(task)
print(result1)
```
